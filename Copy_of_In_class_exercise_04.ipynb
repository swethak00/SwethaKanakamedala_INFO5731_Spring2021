```python
from six.moves import urllib
import pandas as pd

text_array = []
url = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
text_file = urllib.request.urlopen(url)

for line in text_file:
  decoded_data = line.decode("utf-8").replace('\r\n', '')
  if decoded_data: text_array.append(decoded_data)
file = pd.DataFrame (text_array, columns = ['Raw Text'])
print(file)


```

Number of Sentences.

```python
file['Number of Sentences'] = file['Raw Text'].apply(lambda x: len(str(x).split(".")))
file
```

Number of Words

```python
file['Number of Words'] = file['Raw Text'].apply(lambda x: len(str(x).split(" ")))
file
```

Number of Characters

```python
file['Number of Characters'] = file['Raw Text'].str.len()
file
```

Average word length

```python
def avg_word_length(sentence):
  words = sentence.split()
  if len(words) != 0:
    return(sum(len(word) for word in words)/len(words))
  else:
    return None

file['Average Word Length'] = file['Raw Text'].apply(lambda x: avg_word_length(x))
file
```

Number of stopwords


```python
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')

stop = stopwords.words('english')

file['Number of Stopwords'] = file['Raw Text'].apply(lambda x: len([x for x in x.split() if x in stop]))
file
```

Number of special characters

```python
def special_characters(sentence):
  special = 0
  for value in range(len(sentence)):
    if not((sentence[value].isalpha()) and (sentence[value].isdigit())):
        special = special + 1
  return special

file['Number of special characters'] = file['Raw Text'].apply(lambda x: special_characters(x))
file
```

Number of numerics

```python
file['Numbers of numerics'] = file['Raw Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
file
```

Number of uppercase words

```python
file['Number of upper case words'] = file['Raw Text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
file
```

Lower casing

```python
file['Lower Case'] = file['Raw Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
file
```

Punctuation removal

```python
file['Removal of Punctuation'] = file['Lower Case'].str.replace('[^\w\s]','')
file
```

Stopwords removal

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
file['Stopwords Removal'] = file['Removal of Punctuation'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
file
```

Frequent words removal

```python
freq_words = pd.Series(' '.join(file['Stopwords Removal']).split()).value_counts()[:10]
print(freq_words)
freq_words_list = list(freq_words.index)
file['After common words removal'] = file['Stopwords Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_words_list))
file
```

Rare words removal

```python
rare_words = pd.Series(' '.join(file['Stopwords Removal']).split()).value_counts()[-10:]
print(rare_words)
rare_words_list = list(rare_words.index)
file['After Rare Words removal'] = file['After common words removal'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_words_list))
file
```

Spelling correction

```python
from textblob import TextBlob
file['Spelling Correction'] = file['After Rare Words removal'].apply(lambda x: str(TextBlob(x).correct()))
file
```

Tokenization

```python
import nltk
nltk.download('punkt')
file['Tokenization'] = file['Spelling Correction'].apply(lambda x: TextBlob(x).words)
file
```

Stemming

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
file['After Stemming'] = file['Tokenization'].apply(lambda x: " ".join([st.stem(word) for word in x]))
file
```

Lemmatization

```python
from textblob import Word
import nltk
nltk.download('wordnet')

file['After Lemmatization'] = file['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
file
```

Saving to CSV

```python
file.to_csv('cleaned_data.csv',index=False)
```

Advanced Text Processing

```python
term_frequency = (file['After Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
term_frequency.columns = ['words','term_ferquency']
term_frequency
```

Top 10- 1Grams


```python
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

words_sentence = []
for sentence in file['After Lemmatization']:
  words_sentence.append(word_tokenize(sentence))
aftercleaning = [x for x in words_sentence if x != []]
iterations = list(itertools.chain.from_iterable(aftercleaning))
one_grams = ngrams(iterations, 1)
collections.Counter(one_grams).most_common(10)
```

Top 10- 2Grams

```python
bi_grams = nltk.bigrams(iterations)
frequency_dist = nltk.FreqDist(bi_grams)
(frequency_dist.most_common(10))
```

Top 10- 3Grams

```python
tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
frequency_dist.most_common(10)
```
